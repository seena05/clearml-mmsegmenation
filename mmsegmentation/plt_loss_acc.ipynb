```python
import json
res=[]
decoder = json.JSONDecoder()
with open('20220712_result_log.json', 'r') as f:
  line = f.readline()
  while line:
    res.append(decoder.raw_decode(line))
    line = f.readline()

# json_open = open('20220712_result_log.json', 'r')
# json_load = json.load(json_open)

# print(json_load)
# print(res)
losses=[]
accs=[]
for line in res:
  if line[0]['mode']=='train':
    losses.append(line[0]['loss'])
    accs.append(line[0]['aux.acc_seg'])
    # print(line[0]['loss'])
```

```python
import matplotlib.pyplot as plt

x=range(0, len(losses))
plt.plot(x, losses)
plt.show()

x=range(0, len(losses))
plt.plot(x, accs)
plt.show()
```
